```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
unique_users = purchase_data.drop_duplicates(['SN'])
unique_users_count = unique_users.SN.value_counts().sum()
users_data = {'Total Players': [unique_users_count]}
users_df = pd.DataFrame(users_data)
users_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
clean_columns_df = purchase_data.rename(columns={'Purchase ID': 'Purchase_ID', 'Item ID': 'Item_ID', 'Item Name': 'Item_Name'})

unique_items_count = len(clean_columns_df.Item_ID.value_counts())
av_price = clean_columns_df.Price.mean()
total_purchases = clean_columns_df.Purchase_ID.count()
total_rev = clean_columns_df.Price.sum()

summ_data = {'# of Unique Items': [unique_items_count], 'Average Price': [av_price], 'Total Purchases': [total_purchases], 'Total Revenues': [total_rev]}
summ_df = pd.DataFrame(summ_data)
summ_df['Average Price'] = summ_df['Average Price'].map('${:,.2f}'.format)
summ_df['Total Revenues'] = summ_df['Total Revenues'].map('${:,.2f}'.format)
summ_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

gender_count = unique_users[['Gender', 'Purchase ID']].groupby('Gender').count().rename(columns={'Purchase ID': 'Count'})
gender_count[['Percent']] = gender_count[['Count']] * 100/gender_count[['Count']].sum()
gender_count['Percent'] = gender_count['Percent'].map('{:,.2f}%'.format)
gender_count.sort_values(by='Percent', ascending=False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gen_summ_df = purchase_data.groupby('Gender')
gen_df = gen_summ_df.count()
gen_df = gen_df.rename(columns={'Purchase ID': 'Purchase Count'})
gen_df = gen_df.drop(columns=['SN', 'Age', 'Item ID', 'Item Name', 'Price'])

mean_data = gen_summ_df.mean()['Price']
sum_data = gen_summ_df.sum()['Price']
av_total = sum_data/gender_count['Count']
gen_df['Average Price'] = mean_data
gen_df['Total Purchase Value'] = sum_data
gen_df['Avg Total Purchase per Person'] = av_total

gen_df['Average Price'] = gen_df['Average Price'].map('${:,.2f}'.format)
gen_df['Total Purchase Value'] = gen_df['Total Purchase Value'].map('${:,.2f}'.format)
gen_df['Avg Total Purchase per Person'] = gen_df['Avg Total Purchase per Person'].map('${:,.2f}'.format)

gen_df.sort_values(by='Purchase Count', ascending=False)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_df = unique_users
pd.options.mode.chained_assignment = None

age_df["Age Group"] = pd.cut(age_df["Age"], bins, labels=labels)
age_df = age_df.groupby("Age Group")
age_grouped = age_df[["SN"]].count().rename(columns={"SN": "Count"})
age_grouped[["Percentage of Players"]] = age_grouped[["Count"]] * 100 / len(unique_users)
age_grouped[["Percentage of Players"]] = age_grouped["Percentage of Players"].map("{:,.2f}%".format)

age_grouped
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_dup_df = purchase_data
age_dup_df["Age Group"] = pd.cut(age_dup_df["Age"], bins, labels=labels)
age_dup_group = age_dup_df.groupby("Age Group")

age_dup_grouped = age_dup_group[["SN"]].count().rename(columns={"SN": "Purchase Count"})
age_dup_grouped["Average Purchase Price"] = age_dup_group["Price"].mean()
age_dup_grouped["Total Purchase Value"] = age_dup_group["Price"].sum()
age_dup_grouped["Avg Total Purchase per Person"] = age_dup_grouped["Total Purchase Value"] / age_grouped["Count"]

age_dup_grouped["Average Purchase Price"] = age_dup_grouped["Average Purchase Price"].map("${:,.2f}".format)
age_dup_grouped["Total Purchase Value"] = age_dup_grouped["Total Purchase Value"].map("${:,.2f}".format)
age_dup_grouped["Avg Total Purchase per Person"] = age_dup_grouped["Avg Total Purchase per Person"].map("${:,.2f}".format)

age_dup_grouped
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spend_df = purchase_data
top_spend_group = top_spend_df.groupby("SN")

top_spend_grouped = top_spend_group[["Purchase ID"]].count().rename(columns={"Purchase ID": "# of Purchases"})
top_spend_grouped["Average Spent"] = top_spend_group["Price"].mean()
top_spend_grouped["Total Purchase Value"] = top_spend_group["Price"].sum()

top_spend_sorted = top_spend_grouped.sort_values(by="Total Purchase Value", ascending=False)
top_spend_sorted["Average Spent"] = top_spend_grouped["Average Spent"].map("${:,.2f}".format)
top_spend_sorted["Total Purchase Value"] = top_spend_grouped["Total Purchase Value"].map("${:,.2f}".format)

top_spend_sorted.head(6)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = purchase_data
items_group = items_df.groupby("Item ID")

items_grouped = items_group[["SN"]]
items_grouped = items_grouped.count().rename(columns={"SN": "# Purchased"})

# I have no idea how this line of code works past items_group["Item Name"]
# I just kept trying stuff until something worked. Here's my source: https://stackoverflow.com/questions/15222754/groupby-pandas-dataframe-and-select-most-common-value
items_grouped["Item Name"] = items_group["Item Name"].agg(lambda x:x.value_counts().index[0])

items_grouped = items_grouped[["Item Name", "# Purchased"]]
items_grouped["Item Price"] = items_group["Price"].mean().map("${:,.2f}".format)
items_grouped["Total Purchase Value"] = items_group["Price"].sum().map("${:,.2f}".format)
items_pur_sort = items_grouped.sort_values(by="# Purchased", ascending=False)
items_pur_sort
```
